```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Path to the uploaded image file
image_path = '/content/fruits.png'  # Replace with the actual file path

# Read the image
image = cv2.imread(image_path)

# Draw a small green circle in the middle of the image
image_height, image_width = image.shape[:2]
center_coordinates = (int(image_width/2), int(image_height/2))
radius = 20
color = (0, 255, 0)  # BGR color format (Green color)
thickness = -1  # Filled circle
cv2.circle(image, center_coordinates, radius, color, thickness)

# Draw a red line on the top left corner
start_point = (0, 0)
end_point = (50, 50)
color = (0, 0, 255)  # BGR color format (Red color)
thickness = 2
cv2.line(image, start_point, end_point, color, thickness)

# Draw a bright green square on the bottom left corner
square_size = 50
square_top_left = (0, image_height - square_size)
square_bottom_right = (square_size, image_height)
color = (0, 255, 0)  # BGR color format (Green color)
cv2.rectangle(image, square_top_left, square_bottom_right, color, thickness)

# Add a bright green "Yummy" text on the middle right side of the image
text = "Yummy"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (0, 255, 0)  # BGR color format (Green color)
text_thickness = 2
text_size = cv2.getTextSize(text, font, font_scale, text_thickness)[0]
text_width, text_height = text_size
text_origin = (image_width - text_width - 20, int(image_height/2) + int(text_height/2))
cv2.putText(image, text, text_origin, font, font_scale, color, text_thickness, cv2.LINE_AA)

# Convert BGR to RGB for visualization
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with the drawn elements
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Path to the uploaded image file
image_path = '/content/fruits.png'  # Replace with the actual file path

# Read the image
image = cv2.imread(image_path)

# Draw a small green circle in the middle of the image
image_height, image_width = image.shape[:2]
center_coordinates = (int(image_width/2), int(image_height/2))
radius = 20
color = (0, 255, 0)  # BGR color format (Green color)
thickness = -1  # Filled circle
cv2.circle(image, center_coordinates, radius, color, thickness)

# Draw a red line in the top left corner
start_point = (10, 10)
end_point = (70, 70)
color = (0, 0, 255)  # BGR color format (Red color)
thickness = 2
cv2.line(image, start_point, end_point, color, thickness)

# Draw a bright green square in the bottom left corner
square_size = 50
square_top_left = (10, image_height - square_size - 10)
square_bottom_right = (square_top_left[0] + square_size, image_height - 10)
color = (0, 255, 0)  # BGR color format (Green color)
cv2.rectangle(image, square_top_left, square_bottom_right, color, thickness)

# Add a bright green "Yummy" text on the right side of the circle
text = "Yummy"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.7
color = (0, 255, 0)  # BGR color format (Green color)
text_thickness = 2
text_size = cv2.getTextSize(text, font, font_scale, text_thickness)[0]
text_width, text_height = text_size
circle_center_x, circle_center_y = center_coordinates
text_x = circle_center_x + radius + 10  # Distance from the circle
text_y = circle_center_y + text_height // 2
text_origin = (text_x, text_y)
cv2.putText(image, text, text_origin, font, font_scale, color, text_thickness, cv2.LINE_AA)

# Convert BGR to RGB for visualization
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with the drawn elements
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis
plt.show()
```
